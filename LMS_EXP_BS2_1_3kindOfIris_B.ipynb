# Scikit-learn에서 Toy dataset classification해보기
Scikit-learn 패키지는 여러가지 dataset이 제공됩니다. 그 중에서 classification에 적당한 몇 개의 dataset을 machine learning의 다양한 method로 분석해봅시다. 대상 데이터셋은 load_wine, load_digits, load_breast_cancer 셋입니다.

## 


```python
'''
Classification example from Scikit-learn package
Scikit-learn package provide some datasets. For exercise, I'll train some models to classify and discuss about its accuracy.
There are three adequet model for this purpose; load_wine, load_digits, load_breast cancer. There will be 6 steps.
(1)import modules
(2)load dataset
(3)figure out dataset itself
(4)prepare train and test datasets to evaluate models
(5)train 5 models
(6)discuss about models
'''

from sklearn.datasets import load_breast_cancer, load_wine, load_digits
from sklearn.model_selection import train_test_split 
#train_test_split() returns a 4 elements tuple : train(X,Y) and test(X,Y).
from sklearn.metrics import classification_report
#classification_report(y_pred, y_test) shows various values based on the confusion matrix
import pandas as pd
import numpy as np

breast_cancer, wine, digits=load_breast_cancer(), load_wine(), load_digits()
loaded_ds={'breast_cancer':breast_cancer,'wine':wine,'digits':digits}
#for ds in loaded_ds:
#    print(dir(ds.values(ds)/))
```

## 각각의 데이터셋에 모델 적용하기 :
숫자가 너무 많기때문에, 딕셔너리를 이용해 loop를 돌리겠습니다. 처음에는 list를 사용하려고 했는데 숫자 인덱스가 헷갈리기도 하고 list를 수정하는 과정에서 잘못할 가능성이 커서 에러를 수정하다가 그냥 바꿨습니다.


```python
xtrain=dict()
xtest=dict()
ytrain=dict()
ytest=dict()

for dsname,ds in loaded_ds.items():
    data=ds.data
    label=ds.target
    xtrain[dsname],xtest[dsname],ytrain[dsname],ytest[dsname]=train_test_split(data, label, test_size=0.2, random_state=8)

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier, LogisticRegression

models={'DTC':DecisionTreeClassifier(random_state=32), 
        'RFC':RandomForestClassifier(random_state=32), 
        'SVC':svm.SVC(), 
        'SGD':SGDClassifier(), 
        'LR':LogisticRegression()}
for dsname in loaded_ds.keys(): #loop for 3 datasets, as we have splited train/test data assigned to new dictionaries, do not need dataset itself.
    for modelname, model in models.items(): #loop for models.
        ytrn=ytrain[dsname]
        xtrn=xtrain[dsname]
        model.fit(xtrn,ytrn)
        y_pred=model.predict(xtest[dsname])
        print('< DataSet =',dsname, ', Model =', modelname, 'Classification_report >')
        print(classification_report(y_pred,ytest[dsname]))
```

## 데이터셋에 따른 모델의 예측 성능 평가 지표
먼저 classification report에 나오는 값들에 대해 따져봅니다. [wikipedia](https://en.wikipedia.org/wiki/Precision_and_recall)에 따르면 Precision은 False Positive, Recall은 False Negative를 반영합니다.
- 손글씨 숫자 인식의 경우, True Negative를 잘 예측하기가 쉽습니다. 이런 것을 확인하려면 Recall이 균일하게 높게 나타나는지를 봐야 할 것 같습니다. 이 기준에서 SVC가 가장 좋았습니다. 
- 암 진단의 경우 먼저 악성을 양성으로 진단하는 경우를 중요하게 생각해야 하겠습니다. 그런데 0, 1어떤 것이 악성인지 양성인지 분명하지 않습니다. scikit-learn의 description에서 원출처로 링크된 [UCI Machine Learning Library](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic))에서 직접 데이터를 비교해보니 0이 Malignant(악성)입니다. 악성을 양성으로 판단한 것은 class 1에서의 false positive에 해당하므로 class 1의 precision이 가장 중요해 보입니다. 하지만 class 0의 false positive 역시 역시 많은 비용을 발생시키므로 class 0의 precision도 무시하면 안될 것 같습니다. Random Forest Classifier에서 class 1의 precision이 가장 높게 나타났습니다.
- wine 의 경우 데이터 불균형이 심해서 class 0, 2의 recall을 중요하게 봐야 할 것 같습니다. 그런데 RFC가 다른 지표들을 포함해도 가장 정확했습니다.
